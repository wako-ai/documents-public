```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

model = AutoModelForCausalLM.from_pretrained("cyberagent/open-calm-3b", device_map="auto", torch_dtype=torch.float16)
tokenizer = AutoTokenizer.from_pretrained("cyberagent/open-calm-3b")
```

```python
def infer(inputs):
    with torch.no_grad():
        tokens = model.generate(
            **inputs,
            max_new_tokens=64,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            repetition_penalty=1.05,
            pad_token_id=tokenizer.pad_token_id,
        )
    return tokenizer.decode(tokens[0], skip_special_tokens=True)
```

```python
inputs = tokenizer("和光市でのAIの勉強会をしていると、", return_tensors="pt").to(model.device)
infer(inputs)
```

```python
inputs = tokenizer("和光市でのAIの勉強会をしていると、", return_tensors="pt").to(model.device)
infer(inputs)
```
